```python
import matplotlib.pyplot as plt
months=['Jan','Feb','Mar','Apr','May','Jun']
e_consump=[1200,3400,2900,1800,2500,1600]
```

```python
plt.bar(months, e_consump)
plt.title("Energy Consumption over six months")
plt.xlabel("Month")
plt.ylabel("Enery Consumption")
plt.show()
```
