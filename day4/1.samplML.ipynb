```python
import pandas as pd
```

```python
dff=pd.read_csv("weight-height data.csv")
dff
```

```python
df=dff.drop(columns="Gender")
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["Height"],df["Weight"],color='b')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height V/S Weight")
plt.show()
```

```python
X=df[["Height"]]
Y=df["Weight"]
X.shape
Y.shape
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.predict([[73.84701702]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
X_test
```

```python
Y_test
```

```python
accuracy=model.score(X_test,Y_test,y_pred)
accuracy
```
