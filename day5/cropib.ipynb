```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dff=pd.read_csv("/content/sample_data/crop_yield_data.xls")
dff
```

```python
X=dff.drop(["crop_yield"],axis=1)
y=dff[["crop_yield"]]
X.shape,y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler #to match the numbers and weights in DL
scaler=MinMaxScaler()
X_s=scaler.fit_transform(X)
y_s=scaler.fit_transform(y.values.reshape(-1,1)) #y.values -> it changes Series(list) to array.
X_s,y_s
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_s, y_s, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model=Sequential()
model.add(Dense(64,activation="relu",input_dim=X_train.shape[1]))
model.add(Dense(32,activation="relu"))
model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
model.fit(X_train,y_train,epochs=100,batch_size=32)
```

```python
data=np.array([[25,60,50,7,200]])
model.predict(data)
```
